EDA

```python
# monthly analysis by total loan applications
monthly_received=(
    df.sort_values('issue_date')
    .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
    .groupby('month_name',sort=False)['id']
    .count()
    .reset_index(name="total_loan_applications")
)

plt.figure(figsize=(10,5))
plt.fill_between(monthly_received['month_name'],monthly_received['total_loan_applications'],color='lightgreen',alpha=0.5)
plt.plot(monthly_received['month_name'],monthly_received['total_loan_applications'],color='green',linewidth='2')

for i,row in monthly_received.iterrows():
    plt.Text(i, row['total_loan_applications'] + 0.1, f"{row['total_loan_applications']:.2f}",
             ha='center',va='bottom',fontsize='9',rotation=0,color='black')
    

plt.title('Total Loan Applications by month')
plt.xlabel('Month')
plt.ylabel('Loan Applications')
plt.xticks(ticks=range(len(monthly_received)),labels=monthly_received['month_name'],rotation=45)
plt.grid(True,linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# monthly fuded amount

monthly_received=(
    df.sort_values('issue_date')
    .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
    .groupby('month_name',sort=False)['loan_amount']
    .sum()
    .div(1000000)
    .reset_index(name="funded_amount_miliones")
)

plt.figure(figsize=(10,5))
plt.fill_between(monthly_received['month_name'],monthly_received['funded_amount_miliones'],color='lightgreen',alpha=0.5)
plt.plot(monthly_received['month_name'],monthly_received['funded_amount_miliones'],color='green',linewidth='2')

for i,row in monthly_received.iterrows():
    plt.Text(i, row['funded_amount_miliones'] + 0.1, f"{row['funded_amount_miliones']:.2f}",
             ha='center',va='bottom',fontsize='9',rotation=0,color='black')
    

plt.title('Total Funded amount by month')
plt.xlabel('Month')
plt.ylabel('Funded Amount')
plt.xticks(ticks=range(len(monthly_received)),labels=monthly_received['month_name'],rotation=45)
plt.grid(True,linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# monthly received amount

monthly_received=(
    df.sort_values('issue_date')
    .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
    .groupby('month_name',sort=False)['total_payment']
    .sum()
    .div(1000000)
    .reset_index(name="received_amount_miliones")
)

plt.figure(figsize=(10,5))
plt.fill_between(monthly_received['month_name'],monthly_received['received_amount_miliones'],color='lightgreen',alpha=0.5)
plt.plot(monthly_received['month_name'],monthly_received['received_amount_miliones'],color='green',linewidth='2')

for i,row in monthly_received.iterrows():
    plt.Text(i, row['received_amount_miliones'] + 0.1, f"{row['received_amount_miliones']:.2f}",
             ha='center',va='bottom',fontsize='9',rotation=0,color='black')
    

plt.title('Total Received amount by month')
plt.xlabel('Month')
plt.ylabel('Received Amount')
plt.xticks(ticks=range(len(monthly_received)),labels=monthly_received['month_name'],rotation=45)
plt.grid(True,linestyle='--',alpha=0.5)
plt.tight_layout()
plt.show()
```
